## let's start with synodic period

### first load in some standard equations

```python
#super period
def super_period(P1, P2, j, k):
    return 1/np.abs((j/P1) - (k/P2))


#normalized super period to P1
def super_period_norm(P2P1, j, k):
    return 1/np.abs((j) - (k/P2P1))

#normalized super period to P1 and where n now represents the order
#n = 1 means j:j-1 orbit
#n = -1 means j:j+1 orbit
def super_period_norm_order(P2P1, j, n):
    return 1/np.abs((j) - (j-n/P2P1))


#we can solve for j as the nearest resonance to k... 
#since P1/P2 ~ j/k, then j ~ k*(P1/P2) = k/P2P1
def j_nearest_k(P2P1, k):
    return np.round(k/P2P1)

#we can solve for k as the nearest resonance to j... 
#since P1/P2 ~ j/k, then k ~ j/(P1/P2) = k*P2P1
def k_nearest_j(P2P1, k):
    return np.round(k * P2P1)


#aliasing formula
def alias_period(Pttv, m):
    return 1/np.abs((1/Pttv) + m)



#chopping period
def chopping_period(P1, P2):
    return 1/np.abs((1/P1) - (1/P2))
```

### let's define an angle $\theta_{1,1}$ that describes the separation between P1 and P2

```python
# Define a range of Pp/Pt values
Pp_Pt_values = np.linspace(2, 10, 100)

# Calculate the corresponding theta values
theta_values = (2*np.pi / Pp_Pt_values)

# Calculate the periods of the sine function
periods = 2 * np.pi / theta_values

# Plot Pp/Pt vs. Period (T) divided by Pt
plt.figure(figsize=(10, 6))
plt.plot(Pp_Pt_values, periods, label=r'TTV Period from $\theta_{1,1}$', color = colors[1], lw=5, ls='--')
plt.xlabel(r'P$_\mathrm{pert}$/P$_\mathrm{trans}$', fontsize = 27)
plt.ylabel(r'P$_\mathrm{TTVs}$/P$_\mathrm{trans}$', fontsize = 27)
plt.grid(True)
plt.legend(fontsize = 18)
plt.show()
```

### for P2 > 2P1 (ie once alias would kick in), how does this compare to the synodic equations?

```python

    
fig, ax = plt.subplots(1, figsize=[18,9], sharex = True)


#plot the synodic period
p2p1_analytic = np.arange(2/3, 2, .001)
j_analytic = 1
k_analytic = 1
pttv_analytic = super_period_norm(p2p1_analytic, j_analytic, k_analytic)


line1 = ax.plot(p2p1_analytic, pttv_analytic, color = colors[0], lw = 7,
                label = 'synodic period')


# plot aliased synodic period
count = 0
alias_legend_dict = {}
for alias_range in [[2, 10]]:
    p2p1_alias_range_analytic = np.arange(alias_range[0], alias_range[1], .0001)
    j_analytic = 1
    k_analytic = 2
    m = -1
    pttv_alias_range_analytic = alias_period( chopping_period(1, p2p1_alias_range_analytic), m )
    
    ax.plot(p2p1_alias_range_analytic, pttv_alias_range_analytic, colors[0], lw=7,
            label = 'aliased chopping period, m = ' + str(m))
    
    alias_legend_dict['m = ' + str(m)] = 'k'
    count += 1

    
    
#plot theta equation
line2 = ax.plot(Pp_Pt_values, periods, label=r'TTV Period from $\theta_{1,1}$', color = colors[1], lw=5, ls='--')


    
ax.axhline(2, 0, 1, color='k')
plt.text(0.2, 2.1, "Nyquist period", ha='center', va='bottom', fontsize=23, color='black')






ax.set_yscale("log", base=np.e)
ax.set_xscale("log", base=np.e)

ax.set_xlim(0.1, 10)
ax.set_ylim(1, 200)
ax.set_xticks([0.1, 0.5, 1, 5, 10])
ax.set_xticklabels([0.1, 0.5, 1, 5, 10])

ax.set_yticks([1, 5, 10, 50, 100, 200])
ax.set_yticklabels([1, 5, 10, 50, 100, 200])

ax.set_xlabel(r'P$_\mathrm{pert}$/P$_\mathrm{trans}$', fontsize = 27)
ax.set_ylabel(r'P$_\mathrm{TTVs}$/P$_\mathrm{trans}$', fontsize = 27)






ax.grid(True)



# legend entries
lines = line1+line2
labs = [line.get_label() for line in lines]
legend3 = ax.legend(lines, labs, loc=2, fontsize = 23)
ax.add_artist(legend3)  




plt.tight_layout()
plt.savefig('theta_notebook_synodic.pdf')
plt.show()
```

## new general formula?

### $\theta_{j,k} = \frac{k}{j} \, \frac{2\pi}{P_P/P_T}$ 

### ... assuming $k > j$, $P_P > P_T$, and $P_{TTV} < 2P_T$

## How can we interpret $\theta_{j,k}$?


### - $\theta_{j,k}$ is the instantaneous angular separation between the perturbing planet and the theoretical angular separation, $\bar{\theta}_{j,k}$,  of two planets in perfect j:k MMR


### - $\theta_{j,k}$ is defined such it solves the sinusoidal function $TTV_{j,k} = A\sin{\theta_{j,k} t}$. One can then determine the period of such a sine wave via... $P_{TTV_{j,k}} = 2\pi / \theta_{j,k}$

### - $\theta_{j,k}$ is only valid assuming $k > j$, $P_P > P_T$, and $P_{TTV} < 2P_T$ ... the first 2 are just naming conventions, but lets explain the 3rd more carefully...

### - When $P_{TTV} < 2P_T$, the transiting planet is not orbiting quickly enough for the observer to see the true TTV period due to undersampling and the TTV period being less than the Nyquist period. In the context of our equation, we can plug in $P_{TTV} = 2P_T$, and solve for this turn over point... $\frac{P_P}{P_T} > 2\frac{k}{j}$


### - This makes sense as iff the perturbing period is twice as long as the transiting period times the MMR of interest, then the perturbing planet induced TTV will be faster than the Nyquist rate, and will thus be observed at the period predicted by  $\theta_{j,k}$
